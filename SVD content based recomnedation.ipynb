```python
import numpy as np,sys

def content_recomm(user,user_df,film_df):
    u=user_df.iloc[user][1:].values
    print("user experince")
    u_norm=np.linalg.norm(u)
    film_recom=[]
    for row in range(film_df.shape[0]):
        f=film_df.iloc[row][1:].values
        f_name=film_df.index[row]
        f_norm=np.linalg.norm(f)
        s=np.dot(u,f)/(u_norm*f_norm)
        if s>0.8:
            film_recom.append((f_name,s))
#     print("recommendations :",film_recom)
    film_recom=sorted(film_recom,key=lambda x: x[1],reverse=True)
    return film_recom
    
     
# content_recomm(0,user,film)
r1=content_recomm(1,user,film)
print("sorted recomendation",r1)
```
